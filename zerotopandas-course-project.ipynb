## Data Preparation and Cleaning

*We have readed the data with the help of read_csv function and stored it into a new dataframe named covid_df so that it can be further processed and accessed. Then we have checked for missing value and incorrect value by using the .isna, min(), max() methods.*

*Atlast we have added some of the required columns in the dataframe by performing some of the mathematical calculations.*



```python
# Loading the dataset into a data frame using Pandas
covid_df = pd.read_csv('country_wise_latest.csv')
covid_df
```

```python
# Exploring the number of rows & columns
covid_df.shape
```

```python
covid_df.sample(10)
```

```python
# Handle missing, incorrect and invalid data
covid_df[['Confirmed','Deaths','Recovered','Active','New cases','New deaths','New recovered']].min()
```

```python
covid_df[['Confirmed','Deaths','Recovered','Active','New cases','New deaths','New recovered']].max()
```

```python
# All the values are present in the required columns
from IPython.display import display
with pd.option_context('display.max_rows', 200):
    display(covid_df.isna())
```

```python
# Performing any additional steps (parsing dates, creating additional columns, merging multiple dataset etc.)
covid_df['Recovery_rate'] = (covid_df.Recovered / covid_df.Confirmed)*100
covid_df['Death_rate'] = (covid_df.Deaths / covid_df.Confirmed)*100
covid_df
```

```python
world_df = pd.read_csv('worldometer_data.csv')
world_df.drop(columns=['TotalCases','NewCases','TotalDeaths','NewDeaths','TotalRecovered','NewRecovered','ActiveCases','Serious,Critical','Tot Cases/1M pop','Deaths/1M pop','TotalTests','Tests/1M pop','WHO Region'],inplace = True)
world_df
```

```python
merged_df = covid_df.merge(world_df,on='Country')
```

Let's begin by importing`matplotlib.pyplot` and `seaborn`.

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 16
matplotlib.rcParams['figure.figsize'] = (12, 6)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

**Computing the mean, sum, range and other interesting statistics for Dataset columns**

```python
# Computing the mean, sum, range and other interesting statistics for numeric columns
merged_df.describe()
```

```python
merged_df
```

**Here we found the total population of each continent by using the groupby() function of pandas library and then we have plotted the line chart that represents us that which continent has the highest population.**

```python
population_df = merged_df.groupby('Continent')['Population'].sum()
population_df
```

```python
plt.plot(population_df,'o-b')
plt.xlabel('Continents')
plt.ylabel('Population in Billions')
plt.title("Continents-Population");
```

**Here we have drawn a scatter plot chart which tells us the insights of *Death Rate* v/s the *Recovery Rate* of the patients who suffred from the Covid Pandemic.**
**According to the trend that we have seen in this graph *we can conclude that among all the continents Asia* is the richest country in terms of population.**

```python
rate_df = merged_df.groupby('Continent')[['Recovery_rate','Death_rate']].mean()
rate_df
```

```python
plt.title('Death V/S Recovery chart')
sns.scatterplot(x='Recovery_rate', 
                y='Death_rate', 
                hue='Continent',
                s=100,
                data=rate_df);
plt.legend(loc = 'upper left',fontsize=12);
```

```python
# Top 10 countries with highest active cases
active_df = merged_df.sort_values('Active', ascending=False).head(10)
active_df
```

```python
sns.barplot(y='Country',x='Active',data = active_df);
```

**Here we have plotted the horizontal bar chart or we can say a bar plot which describes us the top 10 countries with highest number of active cases.**

```python
population_df = merged_df.groupby('Continent')['Deaths'].sum()
population_df
```

```python
# Pie chart
labels = ['Africa','Asia','Australia','Europe','North America','South America']
sizes = population_df
explode = (0, 0, 0, 0.1,0,0)
plt.pie(sizes,labels=labels,explode=explode);
```

**Here we have plotted the the pie chart by finding the continent which has the hghest death count across all other continents.**

## Asking and Answering Questions

In this secion of the Notebook we have performed multple mathematical operations and functions on the columns of table with the help of *Pandas* library and each functions that we have used in this section are explained above the function used.

### Q1: In the continent region there are plenty of Countries. Find the country of the *Asia continent* which has highest number of active cases.

### Merged_df function
#### *merged_df()* is used to merged two dataframes where one columnand their entries must be common in both.

```python
high_active_df = merged_df[merged_df.Continent=="Asia"]
high_active_df
```

### sort_values and head function
#### *sort_values()* is used to sort the values either in the ascending or in the descending order. *head()* is used to get the number of rows from the top of the dataset or dataframe. By default it returns the top 5 rows.

```python
high = high_active_df.sort_values('Active',ascending=False).head(1)
```

```python
high[['Continent','Country','Active']]
```

### Q2: Find out the top 10 countries which has the Best Recovery Rate all over the world and print the countries and Recovery rate.

```python
top_10_df = merged_df.sort_values('Recovery_rate',ascending=False).head(10)
```

```python
top_10_df
```

```python
top_10_df[['Country','Recovery_rate']]
```

### Q3: Add all the data of new columns like (*New cases,New deaths,New recovered*) in the original data columns (*Confirmed,Deaths,Recovered*) After performing all these operations drop the old columns.

```python
merged_df['Confirmed'] = merged_df['Confirmed'] + merged_df['New cases']
merged_df['Deaths'] = merged_df['Deaths'] + merged_df['New deaths']
merged_df['Recovered'] = merged_df['Recovered'] + merged_df['New recovered']
merged_df
```

### drop function 
#### *drop()* function deletes or removes the columns on the given dataset.

```python
merged_df.drop(columns=['New cases','New deaths','New recovered'],inplace=True)
merged_df
```

### Q4: Perform the aggregate operations on the columns that are amicable and performable for same of Dataset table.

```python
merged_df
```

```python
print("The sum of the columns as per values are :- ")
merged_df[['Confirmed','Deaths','Recovered','Active','Population']].sum()
```

```python
print("The mean of the columns as per values are :- ")
merged_df[['Confirmed','Deaths','Recovered','Active','Recovery_rate','Death_rate','Population']].mean()
```

### Q5: After performing all the operations write back the data to file (like *added columns, manipulated or updtaed data*) in the form of dataset in the new file of extension *csv* .

```python
Final_df = merged_df[['Country','Confirmed','Deaths','Recovered','Active','Recovery_rate','Death_rate','Continent','Population']]
```

```python
Final_df
```

### to_csv Function
#### to_csv() is used to create a new csv file by passing the dataset in a particular dataframe as per the required syntax.

```python
Final_df.to_csv('Final_dataset.csv',index = None)
```
